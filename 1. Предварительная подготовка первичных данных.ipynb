С целью последующего анализа и обработки данных, а также в соотвествтии с условиями поставленной в ВКР задачей, объединим две исходные таблицы в одну по индексу, тип объединения INNER. При использовании данного метода объединяются только те значения, которые можно найти в обеих таблицах, посмотрим объединенный датасет и назовем его ds:

```python
ds = df_bp.merge(df_nup, left_index = True, right_index = True, how = 'inner')
```

Разведочный анализ данных

```python
# Посмотрим на типы признаков:
ds.info()
```

Видим, что в данных отсутствуют пропуски, а все характеристики представлены числовым типом float64. Категориальных признаков нет, как и данных со значением NaN. Преобразования типов не требуется, так как для машинного обучения подходят числовые признаки.

```python
# Поработаем с наличием и количеством уникальных значений в датасете.
ds.nunique()
```

```python
# Видим, что данные с обозначением угла нашивки представены всего двумя уникальными значениями - 0 и 90 градусов. Приведем их к значениям 0 и 1
# Вывыедем количество значений признака, где угол нашивки равен 0 градусов:
ds['Угол нашивки, град'][ds['Угол нашивки, град'] == 0.0].count()
```

```python
#Посмотрим количество данных, где Угол нашивки равен 90 градусов:
ds['Угол нашивки, град'][ds['Угол нашивки, град'] == 90.0].count()
```

```python
# Видим, что таких значений практически поровну. Преобразуем данные к целочисленным 0 и 1:
ds = ds.replace({'Угол нашивки, град': {0.0 : 0, 90.0 : 1}})
ds['Угол нашивки, град'] = ds['Угол нашивки, град'].astype(int)
```

```python
#Переименуем новый столбец, поскольку это уже не градусы, а заданное нами бинарное значение данного признака.
ds = ds.rename(columns={'Угол нашивки, град' : 'Угол нашивки'})
#посмотрим, как изменится датасет:
ds
```

Проверим, не потряны ли данные после преобразования. Видим, что количество данных не изменилось.

```python
ds['Угол нашивки'][ds['Угол нашивки'] == 0.0].count()
```

```python
ds['Угол нашивки'][ds['Угол нашивки'] == 1.0].count()
```
